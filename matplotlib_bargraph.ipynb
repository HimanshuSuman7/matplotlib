```python
# Using csv module
with open("content/stack_overflow.csv") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_counter = Counter()
    for row in csv_reader:
        language_counter.update(row["LanguagesWorkedWith"].split(";"))

print(language_counter)
```

```python
# Selecting 15 most popular languages
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print(popularity)
```

```python
plt.style.use("fivethirtyeight")
plt.barh(languages, popularity)
plt.xlabel("Popularity")
plt.title("Most popular languages")
plt.show()
```

```python
languages.reverse()
popularity.reverse()

plt.style.use("fivethirtyeight")
plt.barh(languages, popularity)
plt.xlabel("Popularity")
plt.title("Most popular languages")
plt.show()
```

```python
# Using pandas module
data = pd.read_csv("content/stack_overflow.csv")
lang_responses = data["LanguagesWorkedWith"]

lang_counter = Counter()

for response in lang_responses:
    lang_counter.update(response.split(";"))

lang_items = list(lang_counter.items())
lang = []
pop = []

for item in lang_items:
    lang.append(item[0])
    pop.append(item[1])

print(lang)
print(pop)
```

```python
var1 = sorted(zip(lang, pop))
lang, pop = map(list, zip(*var1))
print(lang)
print(pop)
```

```python
plt.style.use("ggplot")
plt.barh(lang, pop)
plt.xlabel("Popularity")
plt.title("Most popular languages")
plt.show()
```
