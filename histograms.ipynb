```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcdefaults()
```

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
plt.style.use("fivethirtyeight")
plt.hist(ages, bins=5, edgecolor="black")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.tight_layout()
plt.show()
```

```python
bins = [10, 20, 30, 40, 50, 60]
plt.style.use("fivethirtyeight")
plt.hist(ages, bins=bins, edgecolor="black")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv("content/respondent_age.csv")
data.head()
```

```python
ages = data["Age"]
bins = [x for x in range(0, 101, 5)]
print(bins)
```

```python
plt.style.use("fivethirtyeight")
plt.hist(ages, bins=bins, edgecolor="black")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.hist(ages, bins=bins, edgecolor="black", log=True)
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.tight_layout()
plt.show()
```

```python
data.median(axis=0)
```

```python
median_age = 29
plt.style.use("fivethirtyeight")
plt.hist(ages, bins=bins, edgecolor="black")
plt.axvline(median_age, color="red", linewidth=2, label="Median Age")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.hist(ages, bins=bins, edgecolor="black", log=True)
plt.axvline(median_age, color="red", linewidth=2, label="Median Age")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.tight_layout()
plt.legend()
plt.show()
```
