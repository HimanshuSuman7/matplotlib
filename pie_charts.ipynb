```python
import matplotlib.pyplot as plt
```

```python
plt.rcdefaults()
```

```python
slices = [120, 80]
labels = ["Sixty", "Forty"]
plt.style.use("fivethirtyeight")
plt.pie(slices, labels=labels, wedgeprops={"edgecolor": "black"})
plt.title("Pie-Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [120, 80, 30, 20]
labels = ["elxir", "gold", "silver", "bronze"]
plt.style.use("fivethirtyeight")
plt.pie(slices, labels=labels, wedgeprops={"edgecolor": "black"})
plt.title("Pie-Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 
          'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
plt.style.use("fivethirtyeight")
plt.pie(slices, labels=labels, wedgeprops={"edgecolor": "black"})
plt.title("Pie-Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]     # 10% of the radius
plt.style.use("fivethirtyeight")
plt.pie(slices, labels=labels, explode=explode, wedgeprops={"edgecolor": "black"})
plt.title("Pie-Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]
plt.style.use("fivethirtyeight")
plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, wedgeprops={"edgecolor": "black"})
plt.title("Pie-Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]
plt.style.use("fivethirtyeight")
plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, autopct="%1.1f%%",
        wedgeprops={"edgecolor": "black"})
plt.title("Pie-Chart")
plt.tight_layout()
plt.show()
```
