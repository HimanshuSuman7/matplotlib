```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcdefaults()
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
plt.style.use("seaborn")
plt.scatter(x, y)
plt.tight_layout()
plt.show()
```

```python
plt.style.use("seaborn")
plt.scatter(x, y, s=100, c='green', marker="v")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("seaborn")
plt.scatter(x, y, s=100, c='green', edgecolor="black", linewidth=1, alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.style.use("seaborn")
plt.scatter(x, y, c=colors, edgecolor="black", linewidth=1, alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
plt.style.use("seaborn")
plt.scatter(x, y, c=colors, cmap="Greens", edgecolor="black", linewidth=1, alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
plt.style.use("seaborn")
plt.scatter(x, y, c=colors, cmap="Greens", edgecolor="black", linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label("Satisfaction")
plt.tight_layout()
plt.show()
```

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
plt.style.use("seaborn")
plt.scatter(x, y, s=sizes, c=colors, cmap="Greens", edgecolor="black", linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label("Satisfaction")
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv("content/youtube_trending.csv")
data.head()
```

```python
view_count = data["view_count"]
likes = data["likes"]
ratio = data["ratio"]
```

```python
plt.style.use("seaborn")
plt.scatter(view_count, likes, edgecolor="black", linewidth=1, alpha=0.75)
plt.xlabel("View Count")
plt.ylabel("Likes")
plt.title("Trending YouTube Videos")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("seaborn")
plt.scatter(view_count, likes, edgecolor="black", linewidth=1, alpha=0.75)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("View Count")
plt.ylabel("Likes")
plt.title("Trending YouTube Videos")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("seaborn")
plt.scatter(view_count, likes, c=ratio, cmap="cool", edgecolor="black", linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label("Like/Dislike Ratio")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("View Count")
plt.ylabel("Likes")
plt.title("Trending YouTube Videos")
plt.tight_layout()
plt.show()
```
