```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcdefaults()
```

```python
data = pd.read_csv("content/dev_salary.csv")
```

```python
ages = data["Age"]
dev_salary = data["All_Devs"]
py_salary = data["Python"]
js_salary = data["JavaScript"]
```

```python
plt.plot(ages, dev_salary, linestyle="--", label="All Devs")
plt.plot(ages, py_salary, label="Python")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Report")
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev_salary, linestyle="--", label="All Devs")
plt.plot(ages, py_salary, label="Python")
plt.fill_between(ages, py_salary, alpha=0.25)
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Report")
plt.legend()
plt.show()
```

```python
overall_median = 57287
plt.plot(ages, dev_salary, linestyle="--", label="All Devs")
plt.plot(ages, py_salary, label="Python")
plt.fill_between(ages, py_salary, overall_median, alpha=0.25)
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Report")
plt.legend()
plt.show()
```

```python
overall_median = 57287
plt.plot(ages, dev_salary, linestyle="--", label="All Devs")
plt.plot(ages, py_salary, label="Python")
plt.fill_between(ages, py_salary, overall_median, where=(py_salary > overall_median), interpolate=True, alpha=0.25)
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Report")
plt.legend()
plt.show()
```

```python
overall_median = 57287
plt.plot(ages, dev_salary, linestyle="--", label="All Devs")
plt.plot(ages, py_salary, label="Python")
plt.fill_between(ages, py_salary, overall_median, where=(py_salary > overall_median), interpolate=True, alpha=0.25)
plt.fill_between(ages, py_salary, overall_median, where=(py_salary <= overall_median), interpolate=True, alpha=0.25)
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Report")
plt.legend()
plt.show()
```

```python
plt.plot(ages, dev_salary, linestyle="--", label="All Devs")
plt.plot(ages, py_salary, label="Python")
plt.fill_between(ages, py_salary, dev_salary, where=(py_salary > dev_salary), 
                 interpolate=True, alpha=0.25, label="Above Avg")
plt.fill_between(ages, py_salary, dev_salary, where=(py_salary <= dev_salary), 
                 interpolate=True, alpha=0.25, label="Below Avg")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Report")
plt.legend()
plt.show()
```
