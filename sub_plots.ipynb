```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcdefaults()
```

```python
data = pd.read_csv("content/dev_salary.csv")
data.head()
```

```python
ages = data["Age"]
dev_salary = data["All_Devs"]
py_salary = data["Python"]
js_salary = data["JavaScript"]
```

```python
plt.plot(ages, py_salary, label="Python")
plt.plot(ages, js_salary, label="JavaScript")
plt.plot(ages, dev_salary, linestyle="--", label="All Devs")

plt.legend()
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(ages, py_salary, label="Python")
ax.plot(ages, js_salary, label="JavaScript")
ax.plot(ages, dev_salary, linestyle="--", label="All Devs")

ax.legend()
ax.set_xlabel("Ages")
ax.set_ylabel("Median Salary (USD)")
ax.set_title("Median Salary (USD) by Age")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, py_salary, label="Python")
ax1.plot(ages, js_salary, label="JavaScript")

ax2.plot(ages, dev_salary, linestyle="--", label="All Devs")

ax1.legend()
ax1.set_ylabel("Median Salary (USD)")
ax1.set_title("Median Salary (USD) by Age")
ax1.grid(True)

ax2.legend()
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, py_salary, label="Python")
ax1.plot(ages, js_salary, label="JavaScript")

ax2.plot(ages, dev_salary, linestyle="--", label="All Devs")

ax1.legend()
ax1.set_xlabel("Ages")
ax1.set_ylabel("Median Salary (USD)")
ax1.set_title("Median Salary (USD) by Age")
ax1.grid(True)

ax2.legend()
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")
ax2.set_title("Median Salary (USD) by Age")
ax2.grid(True)

plt.tight_layout()
plt.show()
```
